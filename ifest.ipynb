```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
start = '2020-03-02'
end = '2021-08-21'
giaa = yf.download('GIAA.JK', start = start, end = end, interval='5d')[['Adj Close', 'Volume']]
antm = yf.download('ANTM.JK', start = start, end = end, interval='5d')[['Adj Close', 'Volume']]
kaef = yf.download('KAEF.JK', start = start, end = end, interval='5d')[['Adj Close', 'Volume']]
unvr = yf.download('UNVR.JK', start = start, end = end, interval='5d')[['Adj Close', 'Volume']]
wskt = yf.download('WSKT.JK', start = start, end = end, interval='5d')[['Adj Close', 'Volume']]
```

```python
df_giaa = giaa.diff().dropna()
df_antm = antm.diff().dropna()
df_kaef = kaef.diff().dropna()
df_unvr = unvr.diff().dropna()
df_wskt = wskt.diff().dropna()
```

```python
scaler = MinMaxScaler(feature_range=(-1,1))

scale_giaa = scaler.fit_transform(df_giaa)
sclae_antm = scaler.fit_transform(df_giaa)
scale_kaef = scaler.fit_transform(df_giaa)
scale_unvr = scaler.fit_transform(df_giaa)
scale _wskt = scaler.fit_transform(df_giaa)
```

```python
df_giaa.loc[df_giaa['Adj Close']< 0, 'label'] = 0
df_giaa.loc[df_giaa['Adj Close']>= 0, 'label'] = 1
df_antm.loc[df_antm['Adj Close']< 0, 'label'] = 0
df_antm.loc[df_antm['Adj Close']>= 0, 'label'] = 1
df_kaef.loc[df_kaef['Adj Close']< 0, 'label'] = 0
df_kaef.loc[df_kaef['Adj Close']>= 0, 'label'] = 1
df_unvr.loc[df_unvr['Adj Close']< 0, 'label'] = 0
df_unvr.loc[df_unvr['Adj Close']>= 0, 'label'] = 1
df_wskt.loc[df_wskt['Adj Close']< 0, 'label'] = 0
df_wskt.loc[df_wskt['Adj Close']>= 0, 'label'] = 1
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_giaa.index, df_giaa['Adj Close'])
plt.title("Diferensiasi GIAA")
plt.show()
```
